```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X, y = make_classification(n_samples = 100, n_features = 2, n_informative = 1, n_redundant = 0, n_classes = 2, n_clusters_per_class = 1, random_state = 50, hypercube = False, class_sep = 10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))
plt.scatter(X[:,0],X[:,1],c=y, cmap = 'winter', s= 100)
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(10000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
intercept_, cof = perceptron(X,y)
```

```python
print(intercept_)
print(cof)
```

```python
m = -(cof[0]/cof[1])
b = -(intercept_/cof[1])
```

```python
x_input = np.linspace(-3,3,300)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color = 'red', linewidth = 3)
plt.scatter(X[:,0], X[:,1], c = y, cmap='winter',s =100)
plt.ylim(-3,2)
```
